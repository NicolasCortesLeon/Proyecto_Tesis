```python
import geopandas as gpd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import gaussian_kde
from scipy.ndimage import gaussian_filter

ruta_shp = "../../data/external/shapefiles/archivoscomunaschile/comunas.shp"
gdf = gpd.read_file(ruta_shp)
rm = gdf[gdf['Region'] == "Región Metropolitana de Santiago"]
rm = rm.to_crs("EPSG:4326")

heat_data = df[['latitud', 'longitud', 'precio']].dropna()

fig, ax = plt.subplots(figsize=(16, 14))
ax.set_facecolor('white')

rm.plot(ax=ax, edgecolor='gray', facecolor='#f0f0f0', linewidth=1, alpha=0.5)

# KDE
x, y, z = heat_data['longitud'].values, heat_data['latitud'].values, heat_data['precio'].values
bounds = rm.total_bounds
xx, yy = np.mgrid[bounds[0]:bounds[2]:300j, bounds[1]:bounds[3]:300j]
positions = np.vstack([xx.ravel(), yy.ravel()])

values = np.vstack([x, y])
kernel = gaussian_kde(values, weights=z, bw_method=0.1)  # Más suave
density = np.reshape(kernel(positions).T, xx.shape)

# Aplicar filtro gaussiano adicional para más suavidad
density_smooth = gaussian_filter(density, sigma=2)

# Contour suave
contour = ax.contourf(xx, yy, density_smooth, levels=40, cmap='RdYlBu_r', alpha=0.75)

cbar = plt.colorbar(contour, ax=ax, shrink=0.7)
cbar.set_label('Densidad de Precio', fontsize=14, weight='bold')

rm.plot(ax=ax, edgecolor='black', facecolor='none', linewidth=1.5, alpha=0.4)

# Ajustar
margen = 0.02
ax.set_xlim(bounds[0] - margen * (bounds[2] - bounds[0]), 
            bounds[2] + margen * (bounds[2] - bounds[0]))
ax.set_ylim(bounds[1] - margen * (bounds[3] - bounds[1]), 
            bounds[3] + margen * (bounds[3] - bounds[1]))
ax.set_aspect('equal')

plt.title('Mapa de Densidad de Precios - RM', fontsize=18, weight='bold', pad=20)
plt.tight_layout()
plt.show()
```
