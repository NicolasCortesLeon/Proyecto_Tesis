```python
fig, axes = plt.subplots(1, 2, figsize=(15, 6))

max_freq_lectura = max(axes[0].hist(df['simce 4to básico lectura 2022'], bins=50)[0])
max_freq_matematicas = max(axes[1].hist(df['simce 4to básico matemáticas 2022'], bins=50)[0])
max_freq = max(max_freq_lectura, max_freq_matematicas)

axes[0].clear()
axes[1].clear()

axes[0].hist(df['simce 4to básico lectura 2022'], bins=50, edgecolor='black', alpha=0.7)
axes[0].set_xlabel('SIMCE Lectura 2022')
axes[0].set_ylabel('Frecuencia')
axes[0].set_title('Distribución SIMCE 4to Básico Lectura 2022')
axes[0].set_ylim(0, max_freq * 1.05)  
axes[0].grid(True, alpha=0.3)

axes[1].hist(df['simce 4to básico matemáticas 2022'], bins=50, edgecolor='black', alpha=0.7)
axes[1].set_xlabel('SIMCE Matemáticas 2022')
axes[1].set_ylabel('Frecuencia')
axes[1].set_title('Distribución SIMCE 4to Básico Matemáticas 2022')
axes[1].set_ylim(0, max_freq * 1.05) 
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

Varios gráficos se pueden detectar la forma de dispersión de los puntos en los scatterplots heterocedasticidad, por lo que aunque en modelos como rf, xgboost y lgbm no tienen problema con esto, para modelos como regresión lineal, svm, o redes neuronales, la heterocedasticidad es ciertamente un problema que es preferible que sea eliminado.
Por otro lado podemos ver que los dormitorios no presentan una correlación lineal muy fuerte como si lo presentan los baños, pero los baños luego de 5, pierden un poco de su valor agregado.
Por el lado de SIMCE de matemáticas y lenguaje se ve que presentan una correlación lineal fuerte con el precio.
Aunque si vemos ahora la distancia al hospital más cercano y a la estación de metro más cercano, podemos ver que no existe una correlación lineal muy fuerte, pero como vimos anteriormente, esto no quiere decir que no nos pueda aportar poder predictivo si usamos modelos que pueden encontrar relaciones no lineales.

```python
fig, axes = plt.subplots(5, 4, figsize=(20, 25))
axes = axes.ravel()

columnas = [col for col in df.columns if col != 'precio']

for idx, columna in enumerate(columnas):
    axes[idx].scatter(df[columna], df['precio'], alpha=0.5, s=10)
    axes[idx].set_xlabel(columna, fontsize=9)
    axes[idx].set_ylabel('Precio', fontsize=9)
    axes[idx].set_title(f'Precio vs {columna}', fontsize=10)
    axes[idx].grid(True, alpha=0.3)
for idx in range(len(columnas), len(axes)):

    axes[idx].set_visible(False)

plt.tight_layout()
plt.show()
```

Una cosa que me gustaría recalcar aquí, es que como podemos ver en esta matriz de correlación, las 2 variables como la distancia al hospital más cercano y la distancia a la estación de metro más cercano no tienen una correlación lineal muy fuerte con el precio, porque de hecho son la variable penúltima y antepenúltima en términos de su fuerza de correlación lineal, ¿pero por qué en feature importance estas variables tienen más peso? De hecho en el feature importance promedio son la 8va y 10ma, y es que esto es lo bueno de los modelos de machine learning, capturan correlaciones no lineales entre estas 2 variables con el precio y mejoran la predicción del precio.

```python
correlaciones_precio = df.corr()['precio'].abs().sort_values(ascending=False)

orden_variables = correlaciones_precio.index.tolist()

correlacion_matrix_ordenada = df[orden_variables].corr()

plt.figure(figsize=(16, 14))

sns.heatmap(correlacion_matrix_ordenada, 
            annot=True,  
            fmt='.2f',  
            cmap='coolwarm',  
            center=0,  
            square=True, 
            linewidths=0.5,
            cbar_kws={"shrink": 0.8})

plt.title('Matriz de Correlación - Ordenada por Correlación Absoluta con Precio', fontsize=16, fontweight='bold', pad=20)
plt.xticks(rotation=45, ha='right', fontsize=9)
plt.yticks(rotation=0, fontsize=9)
plt.tight_layout()
plt.show()
```
