Una cosa que me gustaría recalcar aquí, es que como podemos ver en esta matriz de correlación, las 2 variables como la distancia al hospital más cercano y la distancia a la estación de metro más cercano no tienen una correlación lineal muy fuerte con el precio, porque de hecho son la variable penúltima y antepenúltima en términos de su fuerza de correlación lineal, ¿pero por qué en feature importance estas variables tienen más peso? De hecho en el feature importance promedio son la 8va y 10ma, y es que esto es lo bueno de los modelos de machine learning, capturan correlaciones no lineales entre estas 2 variables con el precio y mejoran la predicción del precio.

```python
import seaborn as sns

correlaciones_precio = df.corr()['precio'].abs().sort_values(ascending=False)

orden_variables = correlaciones_precio.index.tolist()

correlacion_matrix_ordenada = df[orden_variables].corr()

plt.figure(figsize=(16, 14))

sns.heatmap(correlacion_matrix_ordenada, 
            annot=True,  
            fmt='.2f',  
            cmap='coolwarm',  
            center=0,  
            square=True, 
            linewidths=0.5,
            cbar_kws={"shrink": 0.8})

plt.title('Matriz de Correlación - Ordenada por Correlación Absoluta con Precio', fontsize=16, fontweight='bold', pad=20)
plt.xticks(rotation=45, ha='right', fontsize=9)
plt.yticks(rotation=0, fontsize=9)
plt.tight_layout()
plt.show()
```

```python
import geopandas as gpd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import gaussian_kde
from scipy.ndimage import gaussian_filter

ruta_shp = "../../data/external/shapefiles/archivoscomunaschile/comunas.shp"
gdf = gpd.read_file(ruta_shp)
rm = gdf[gdf['Region'] == "Región Metropolitana de Santiago"]
rm = rm.to_crs("EPSG:4326")

heat_data = df[['latitud', 'longitud', 'precio']].dropna()

fig, ax = plt.subplots(figsize=(16, 14))
ax.set_facecolor('white')

rm.plot(ax=ax, edgecolor='gray', facecolor='#f0f0f0', linewidth=1, alpha=0.5)

# KDE
x, y, z = heat_data['longitud'].values, heat_data['latitud'].values, heat_data['precio'].values
bounds = rm.total_bounds
xx, yy = np.mgrid[bounds[0]:bounds[2]:300j, bounds[1]:bounds[3]:300j]
positions = np.vstack([xx.ravel(), yy.ravel()])

values = np.vstack([x, y])
kernel = gaussian_kde(values, weights=z, bw_method=0.1)  # Más suave
density = np.reshape(kernel(positions).T, xx.shape)

# Aplicar filtro gaussiano adicional para más suavidad
density_smooth = gaussian_filter(density, sigma=2)

# Contour suave
contour = ax.contourf(xx, yy, density_smooth, levels=40, cmap='RdYlBu_r', alpha=0.75)

cbar = plt.colorbar(contour, ax=ax, shrink=0.7)
cbar.set_label('Densidad de Precio', fontsize=14, weight='bold')

rm.plot(ax=ax, edgecolor='black', facecolor='none', linewidth=1.5, alpha=0.4)

# Ajustar
margen = 0.02
ax.set_xlim(bounds[0] - margen * (bounds[2] - bounds[0]), 
            bounds[2] + margen * (bounds[2] - bounds[0]))
ax.set_ylim(bounds[1] - margen * (bounds[3] - bounds[1]), 
            bounds[3] + margen * (bounds[3] - bounds[1]))
ax.set_aspect('equal')

plt.title('Mapa de Densidad de Precios - RM', fontsize=18, weight='bold', pad=20)
plt.tight_layout()
plt.show()
```
