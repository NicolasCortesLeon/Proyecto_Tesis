```python
import geopandas as gpd

archivo = '../../data/external/shapefiles/layer_establecimientos_de_salud/layer_establecimientos_de_salud_20240712091725.shp'

gdf = gpd.read_file(archivo)

print(gdf.head())

print("/nColumnas disponibles:")
print(gdf.columns)

if 'categoria' in gdf.columns:
    print("/nCategorías únicas de establecimientos:")
    print(gdf['categoria'].unique())
elif 'tipo' in gdf.columns:
    print("/nTipos únicos de establecimientos:")
    print(gdf['tipo'].unique())
```

```python
hospitales_rm = gdf[
    (gdf['TIPO'].str.contains('HOSPITAL', case=False, na=False)) &
    (gdf['PERTENENCI'].str.contains('SISTEMA NACIONAL', case=False, na=False)) &
    (gdf['REGION'].str.upper() == 'METROPOLITANA DE SANTIAGO')
]

hospitales_rm[['TIPO', 'geometry', 'NOMBRE']]
```

```python
hospitales_rm_publicos = hospitales_rm[
    hospitales_rm['PERTENENCI'].str.startswith('PERTENECIENTE AL SISTEMA NACIONAL', na=False)
].copy()

hospitales_rm_publicos['longitud hospital'] = hospitales_rm_publicos.geometry.x
hospitales_rm_publicos['latitud hospital'] = hospitales_rm_publicos.geometry.y

print(hospitales_rm_publicos[[ 'latitud hospital', 'longitud hospital', 'NOMBRE']])
```

```python
import matplotlib.pyplot as plt
from shapely.geometry import Point

ruta_shp = "../../data/external/shapefiles/archivoscomunaschile/comunas.shp"
gdf = gpd.read_file(ruta_shp)

rm = gdf[gdf['Region'] == "Región Metropolitana de Santiago"]

geometry = [Point(xy) for xy in zip(hospitales_rm_publicos['longitud hospital'], 
                                     hospitales_rm_publicos['latitud hospital'])]

gdf_hospitales = gpd.GeoDataFrame(
    hospitales_rm_publicos, 
    geometry=geometry,
    crs='EPSG:4326'
)

gdf_hospitales = gdf_hospitales.to_crs(rm.crs)

fig, ax = plt.subplots(figsize=(12, 12))

rm.plot(ax=ax, edgecolor='black', facecolor='lightblue', linewidth=0.8, alpha=0.6)

gdf_hospitales.plot(ax=ax, 
                    color='red', 
                    markersize=100, 
                    alpha=0.8,
                    edgecolor='darkred',
                    linewidth=1.5,
                    label='Hospitales Públicos')

for idx, row in rm.iterrows():
    x, y = row['geometry'].centroid.x, row['geometry'].centroid.y
    plt.text(x, y, row['Comuna'], fontsize=7, ha='center', alpha=0.7)

plt.title("Hospitales Públicos - Región Metropolitana", fontsize=14, fontweight='bold')
plt.legend(loc='upper right')
plt.axis('off') 
plt.tight_layout()
plt.show()
```

```python
import folium
import geopandas as gpd
from shapely.geometry import Point

geometry = [Point(xy) for xy in zip(hospitales_rm_publicos['longitud hospital'], 
                                     hospitales_rm_publicos['latitud hospital'])]

gdf_hospitales = gpd.GeoDataFrame(
    hospitales_rm_publicos, 
    geometry=geometry,
    crs='EPSG:4326'
)

gdf_hospitales['lon'] = gdf_hospitales.geometry.x
gdf_hospitales['lat'] = gdf_hospitales.geometry.y

mapa = folium.Map(
    location=[-33.45, -70.65], 
    zoom_start=11,
    tiles='OpenStreetMap' 
)

for _, row in gdf_hospitales.iterrows():
    folium.Marker(
        location=[row['lat'], row['lon']],
        popup=f"<b>{row['NOMBRE']}</b>",
        tooltip=row['NOMBRE'],
        icon=folium.Icon(color='red', icon='plus', prefix='fa')
    ).add_to(mapa)

mapa
```
